```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV file

file="Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read csv file with pandas

svyclean_data_df=pd.read_csv(file, encoding = 'iso-8859-1', low_memory=False)
svyclean_data_df.head()
```

```python
# Inspect all columns of dataframe

svyclean_data_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_svyclean_df=svyclean_data_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_svyclean_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced_svyclean_df=reduced_svyclean_df.replace({0.0: "No", 1.0: "Yes"})
reduced_svyclean_df.head()
```

```python
# Calculate total number of respondents in survey

respondents=len(reduced_svyclean_df)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
reduced_svyclean_df[['AttendedBootcamp','BootcampFinish','BootcampFullJobAfter','BootcampLoanYesNo','BootcampPostSalary']]
```

```python
#  Create a new df for "bootcamp_attended" with the columns.
boot_camp_attendees_df=reduced_svyclean_df[reduced_svyclean_df.AttendedBootcamp=="Yes"]
boot_camp_attendees_df.head()
```

```python
# Calculate average age of attendees
avg_age_attendees= boot_camp_attendees_df['Age'].mean()
avg_age_attendees
```

```python
# Calculate how many people attended a bootcamp
boot_camp_attendees_df['AttendedBootcamp'].count()
```

```python
#calculate length of rows to find total attendies

total_attendees=len(boot_camp_attendees_df)
total_attendees
```

```python
# School degrees

boot_camp_attendees_df['SchoolDegree'].unique()
```

```python
# Calculate how many attendees hold degrees

attendees_schl_degree=boot_camp_attendees_df['SchoolDegree'].count()
attendees_schl_degree
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

boot_camp_attendees_df.Gender.value_counts()
```

```python
# Calculate percentage of respondents who attended a bootcamp

percent_attendees=total_attendees/respondents*100
percent_attendees
```

```python
# Calculate percentage of respondents belonging to each gender

percent_gender=reduced_svyclean_df.Gender.value_counts()/respondents*100
percent_gender
```

```python
# Calculate percentage of attendees with a school degree

percent_schl_degree= attendees_schl_degree/total_attendees*100
percent_schl_degree
```

```python
# Calculate average post-bootcamp salary

Bootcamp_avg_salary=boot_camp_attendees_df['BootcampPostSalary'].mean()
Bootcamp_avg_salary
```

```python
# Create a new table consolodating above calculations

Summary_Bootcamp=pd.DataFrame({
    'Total_Respondents':[respondents],
    'Total_attended_Bcamp':[total_attendees],
    'Percent_attendees':[percent_attendees],
    'Avg_age':[avg_age_attendees],
    'School_degree':[attendees_schl_degree],
    'Percent_school_degree':[percent_schl_degree],
    'Avg_Salary':[Bootcamp_avg_salary],
    'Gender':[percent_gender]
})

Summary_Bootcamp = Summary_Bootcamp.round(2)

Summary_Bootcamp
```

```python
# Improve formatting before outputting spreadsheet

Summary_Bootcamp["Percent_attendees"] =Summary_Bootcamp['Percent_attendees']
Summary_Bootcamp['Avg_age'] =Summary_Bootcamp['Avg_age']
Summary_Bootcamp['School_degree'] =Summary_Bootcamp['School_degree']
Summary_Bootcamp['Percent_school_degree'] =Summary_Bootcamp['Percent_school_degree'].map("{0:,.2f}%".format)
Summary_Bootcamp['Avg_Salary'] =Summary_Bootcamp['Avg_Salary'].map("${0:,.2f}".format)

Summary_Bootcamp
```

```python
# Export to Excel

Summary_Bootcamp.to_excel('output/Bootcamp_partA.xlsx')
```
